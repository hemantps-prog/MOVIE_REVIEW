```python
#document_words_tuple contains the list of tuple of Documents text and its category

document_words_tuple=[]
for category in movie_reviews.categories():
    for fileids in movie_reviews.fileids(category):
        document_words_tuple.append((movie_reviews.words(fileids),category))
    
document_words_tuple
    
```

```python
import random 
random.shuffle(document_words_tuple)

document_words_tuple[0:5]
```

```python
#Importing length and stopwords

from nltk.corpus import stopwords
import string
stopwords=set(stopwords.words('english'))
stopwords=list(stopwords)
stopwords=list(string.punctuation)+stopwords
stopwords
```

```python
def simple_words(word):
    if word.startswith('J'):
        return wordnet.ADJ
    elif word.startswith('N'):
        return wordnet.NOUN
    elif word.startswith('R'):
        return wordnet.ADJ
    elif word.startswith('V'):
        return wordnet.VERB
    else:
        return wordnet.NOUN
    
print(simple_words("JJJ"))
    
    
```

```python
def clean_words(words):
    clean_words=[]
    for w in words:
        if w.lower() not in stopwords:
            lemmatizer=WordNetLemmatizer()
            pos=pos_tag([w])
            clean_words.append(lemmatizer.lemmatize(w.lower(),pos=simple_words(pos[0][1])))
    return clean_words
            
            
```

```python


all_documents=[(clean_words(doc_name),category) for doc_name,category in document_words_tuple]
#Building Featire set
# all_words=[]
# for doc in document_words_tuple:
#     all_words+=list(clean_words(doc[0]))

# all_words

# print(type(document_words_tuple[0][0]))
```

```python
all_words=[]
for doc in all_documents:
    all_words+=doc[0]

all_words
```

```python
#Splitting Data into training and testing 

training_doc=all_documents[0:1500]
testing_doc=all_documents[1500:]

training_doc[0:5]
```

```python
import nltk
frq=nltk.FreqDist(all_words)
most_common_words=frq.most_common(3000)
```

```python
features=[i for i,j in most_common_words]
```

```python


def build_features_dict(words):
    current_features={}
    wordset=set(words)
    for w in features:
        current_features[w]=w in wordset
    return current_features


```

```python
training_doc_dictionary=[(build_features_dict(doc),category) for doc,category in training_doc]
testing_doc_dictionary=[(build_features_dict(doc),category) for doc,category in testing_doc]
```

### USING NAIVE BAYES CLASSIFIER

```python
from nltk import NaiveBayesClassifier

clf=NaiveBayesClassifier.train(training_doc_dictionary)

```

```python
nltk.classify.accuracy(clf,testing_doc_dictionary)
```
